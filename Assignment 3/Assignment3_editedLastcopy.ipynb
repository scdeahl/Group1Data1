## Exercise 1: Loading the data:

- Load the `goodreads.csv` file into Python

- Change the column names to `["rating", 'review_count', 'isbn', 'booktype','author_url', 'year', 'genre_urls', 'dir','rating_count', 'name']`



```python
# This imports the pandas library
import pandas as pd
# use pd.read_csv() to read the CSV file located at "data/goodreads.csv" 
# The header=None parameter indicates that the CSV file does not have a header row, so pandas will not use the first row as column names
# Instead, it will assign default integer column names. These we can easily change later
gr = pd.read_csv('data/goodreads.csv', header=None)
# rename columns names 
gr.columns = ["rating", 'review_count', 'isbn', 'booktype', 'author_url', 'year', 'genre_urls', 'dir', 'rating_count', 'name']
#display new column names and first and last 5 rows
gr.head()
```

## Exercise 2: Subsetting the data

- Subset the data by creating new dataframe only with `["rating", 'isbn', 'author_url', 'year', 'genre_urls', 'name']`

```python
# we create a new DataFrame gr_subset from an existing DataFrame gr.
# The list ["rating", 'isbn', 'author_url', 'year', 'genre_urls', 'name'] specifies the column names we want to extract from the original DataFrame gr
gr_subset = gr[["rating", 'isbn', 'author_url', 'year', 'genre_urls', 'name']]

# Just a quick look on the newly created dataframe
display(gr_subset)
```

## Exercise 3: Manage Missing Data
We’ve got a number of ways in general of dealing with missing data. These involve

1. Dropping off cases (or rows) in the data with any missing variables
2. Excluding variables in the data with any missing data 
3. Selectively choosing indicators with only a limited amount of missing data
4. Replacing missing variables with averages, or other representative values
5. Creating a separate model to predict missing data

- Count the missing values in each column
- Manage the missing values (delete or replace values or leave them as they are) and briefly explain your choice for each column


```python
# In order to count how many missing values (NaN) are present in each column.
# We check each element in the DataFrame gr_subset to see if it contains a null (or NaN) value using the function .isnull()
# .isnull() returns a DataFrame of the same shape as gr_subset, with True for null values and False for non-null values.
# .sum(): sums the True values (which are treated as 1) column-wise, giving the total count of missing values in each column of the DataFrame.
print("Missing per column:\n", gr_subset.isnull().sum())
```

```python
# A quick check of the NaN values in the 'name' column reveals that the 'rating', 'name' and 'author_url' columns share the same rows with missing values (NaN).
gr_subset.loc[gr_subset.name.isnull()]
```

```python
# These 2 rows will be deleted from our new dataframe because we don't need them to our dataframe.
# By dropping these rows we handled the NaN values for those 3 columns 'rating', 'name' and 'author_url'
gr_subset = gr_subset.dropna(subset=['rating'], inplace=False)

# Confirmation of deletion
gr_subset.loc[gr_subset.name.isnull()]
```

```python
#We decided to keep the rest NaN values and replace them with "unknown" beacuse the other columns give us important 
# information and one misssing value in a column does not mean that the whole row is not essential.
gr_subset= gr_subset.fillna("unknown")

# Display the new dataframe to see changes
display(gr_subset)
```

## Exercise 4: Shape the data
- Parse the `author_url` to create new column named `author`
- Sort the data by putting higher rates go first. If there are overlapping rates, try to put earlier years go first.
- **(Stretch Goal)** Examine how many books were published at each year and find lowest, highest rate of each year. 

```python
#writing a function, that clears the url from everything except the name of the author
def get_author(url):

    name = url.split('.')[-1]   # taking the part after the last dot
    name = name.split('_')      # spliting the text into a list, divided by '_'
    name = ' '.join(name)       # forming a string from the items of the list
    return name

#creating a new column with the function
gr_subset['author']=gr_subset.author_url.map(get_author)
display(gr_subset)
```

```python
#sorting the values:
books_reduced_by_authors = gr_subset.sort_values(by=['rating', 'year'], ascending=[False, True])
books_reduced_by_authors.head()
```

```python
display('Books published in the year: ', gr_subset.groupby(['year']).size())
display('Lowest rate for a book in the year: ', gr_subset.groupby(['year']).min(['rating']))
display('Highest rate for a book in the year: ', gr_subset.groupby(['year']).max(['rating']))
```

- Create a scatter plot showing the relationship between these two features.
- Based on the plot and the correlation, provide a brief interpretation of the relationship.

### Python Tools: Use pandas and numpy for correlation, and matplotlib or seaborn for the scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

cleaned_goodreads=pd.read_csv("data/cleaned-goodreads.csv", header=0)

correlation = cleaned_goodreads['rating'].corr(cleaned_goodreads['review_count'], method='pearson') #standar method 'pearson'
print('Correlation between rating and review: \n', correlation)
```

Pearson correlation is close to 0, what means that there is no linear correlation between the the datasets. In other words, having more reviews doesn’t consistently lead to higher or lower ratings.

```python
_=sns.scatterplot(cleaned_goodreads.loc[:, ["rating", 'review_count']], x="review_count", y="rating")
# Title of the plot
plt.title('Review Count vs. Rating') 
```

"There is no observable correlation between the number of reviews and ratings, as shown in the chart. 
The data suggests that the number of reviews does not have a linear effect on the ratings. Most books have fewer than 20,000 reviews, 
and their ratings tend to stay within the range of 3 to 5. Both books with low and high ratings generally have low review counts. We can see the books with low 
and high ratings are those with low reviews. The books that has a avergae rating close to the mean/median usually have higher reviews. Based on these observations linear regression 
is not the exact function to describe the data, so no Pearson (linear) correlation can be detected
